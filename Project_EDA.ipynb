```python
# Filter rows where 'Recipient id_duplicate' column is not null
filtered_df = df[df['Recipient id_duplicate'].notna()]
filtered_df['address_state_city_street'] = filtered_df['address_state_city_street'] + '-' + filtered_df['Recipient id']
filtered_df['Recipient id'] = filtered_df['Recipient id_duplicate']
filtered_df.drop(columns=['Recipient id_duplicate'], inplace=True)
```

```python
# Assign the filtered data back to the original DataFrame 'df'
df.loc[df['Recipient id_duplicate'].notna(), :] = filtered_df
```
